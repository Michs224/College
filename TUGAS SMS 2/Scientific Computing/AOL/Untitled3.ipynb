```python
##COS TAYLOR

import math
def func_cos(x, n):
    cos_approx = 0
    for i in range(n):
        coef = (-1)**i
        num = x**(2*i)
        denom = math.factorial(2*i)
        cos_approx += ( coef ) * ( (num)/(denom) )
    
    return cos_approx

angle_rad = (math.radians(90))
print(angle_rad)
out = func_cos(angle_rad,4)
print(out)
```

```python
#SIN TAYLOR

import math
def func_sin(x, n):
    sin_approx = 0
    for i in range(n):
        coef = (-1)**i
        num = x**(2*i+1)
        denom = math.factorial(2*i+1)
        sin_approx += ( coef ) * ( (num)/(denom) )
    
    return sin_approx

angle_rad = (math.radians(90))
print(angle_rad)
out = func_sin(angle_rad,4)
print(out)
```

```python
##SINXCOSX TAYLOR
import math

def taylor_series_sin_cos(x, n):
    result = 0
    for i in range(n):
        coefficient = ((-1) ** i) / math.factorial(2 * i + 1)
        term = coefficient * (math.sin(x) ** (2 * i + 1)) * (math.cos(x) ** (2 * i + 1))
        result += term
    return result

angle_rad = (math.radians(90))
out = func_sin(angle_rad,4)
print(out)
```

```python
import math
def func_cos(x, n):
    cos_approx = 0
    for i in range(n):
        coef = (-1)**i
        num = x**(2*i)
        denom = math.factorial(2*i)
        cos_approx += ( coef ) * ( (num)/(denom) )
    
    return cos_approx

angle_rad = (math.radians(45))
print(angle_rad)
out = func_cos(angle_rad,4)
print(out)

def truncation_error(x, terms):
  """
  Calculates the truncation error of the Taylor series for cos(x) with x = 0 and terms = 4th.

  Args:
    x: The point at which to evaluate the Taylor series.
    terms: The number of terms to evaluate.

  Returns:
    The truncation error of the Taylor series.
  """
  f_fourth_derivative = -math.cos(x)
  f_fourth_derivative_at_zero = -1
  truncation_error = f_fourth_derivative_at_zero * (x ** (terms + 1)) / math.factorial(terms + 1)
  return truncation_error

truncation_error = truncation_error(math.radians(45), 4)
print(truncation_error)

##the approximation cos(/4)  1 - ^2/32 is accurate to within ^4/256.
```

```python
f = lambda x : x**3-0.3*x**2-8.56*x+8.448
a = 0; b = np.pi*2; N = 20
n = 10 # Use n*N+1 points to plot the function smoothly

x = np.linspace(a,b,N+1)
y = f(x)

X = np.linspace(a,b,n*N+1)
Y = f(X)

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.plot(X,Y,'b')
x_left = x[:-1] # Left endpoints
y_left = y[:-1]
plt.plot(x_left,y_left,'b.',markersize=10)
plt.bar(x_left,y_left,width=(b-a)/N,alpha=0.2,align='edge',edgecolor='b')
plt.title('Left Riemann Sum, N = {}'.format(N))

plt.subplot(1,3,2)
plt.plot(X,Y,'b')
x_mid = (x[:-1] + x[1:])/2 # Midpoints
y_mid = f(x_mid)
plt.plot(x_mid,y_mid,'b.',markersize=10)
plt.bar(x_mid,y_mid,width=(b-a)/N,alpha=0.2,edgecolor='b')
plt.title('Midpoint Riemann Sum, N = {}'.format(N))

plt.subplot(1,3,3)
plt.plot(X,Y,'b')
x_right = x[1:] # Left endpoints
y_right = y[1:]
plt.plot(x_right,y_right,'b.',markersize=10)
plt.bar(x_right,y_right,width=-(b-a)/N,alpha=0.2,align='edge',edgecolor='b')
plt.title('Right Riemann Sum, N = {}'.format(N))

plt.show()
```

```python
dx = (b-a)/N
x_left = np.linspace(a,b-dx,N)
x_midpoint = np.linspace(dx/2,b - dx/2,N)
x_right = np.linspace(dx,b,N)

print("Partition with",N,"subintervals.")
left_riemann_sum = np.sum(f(x_left) * dx)
print("Left Riemann Sum:",left_riemann_sum)

midpoint_riemann_sum = np.sum(f(x_midpoint) * dx)
print("Midpoint Riemann Sum:",midpoint_riemann_sum)

right_riemann_sum = np.sum(f(x_right) * dx)
print("Right Riemann Sum:",right_riemann_sum)
```

```python
# Trapezoidal Method

# Define function to integrate
def f(x):
    return x**3-0.3*x**2-8.56*x+8.448

# Implementing trapezoidal method
def trapezoidal(x0,xn,n):
    # calculating step size
    h = (xn - x0) / n
    
    # Finding sum 
    integration = f(x0) + f(xn)
    
    for i in range(1,n):
        k = x0 + i*h
        integration = integration + 2 * f(k)
    
    # Finding final integration value
    integration = integration * h/2
    
    return integration
    
# Input section
lower_limit = 0
upper_limit = np.pi*2
sub_interval = 20

# Call trapezoidal() method and get result
result = trapezoidal(lower_limit, upper_limit, sub_interval)
print("Integration result by Trapezoidal method is: %0.6f" % (result) )
```

```python
# Simpson's 1/3 Rule

# Define function to integrate
def f(x):
    return x**3-0.3*x**2-8.56*x+8.448

# Implementing Simpson's 1/3 
def simpson13(x0,xn,n):
    # calculating step size
    h = (xn - x0) / n
    
    # Finding sum 
    integration = f(x0) + f(xn)
    
    for i in range(1,n):
        k = x0 + i*h
        
        if i%2 == 0:
            integration = integration + 2 * f(k)
        else:
            integration = integration + 4 * f(k)
    
    # Finding final integration value
    integration = integration * h/3
    
    return integration
    
# Input section
lower_limit = 0
upper_limit = np.pi*2
sub_interval = 20

# Call trapezoidal() method and get result
result = simpson13(lower_limit, upper_limit, sub_interval)
print("Integration result by Simpson's 1/3 method is: %0.6f" % (result) )
```

```python
# Simpson's 3/8 Rule

# Define function to integrate
def f(x):
    return x**3-0.3*x**2-8.56*x+8.448

# Implementing Simpson's 3/8
def simpson38(x0,xn,n):
    # calculating step size
    h = (xn - x0) / n
    
    # Finding sum 
    integration = f(x0) + f(xn)
    
    for i in range(1,n):
        k = x0 + i*h
        
        if i%3 == 0:
            integration = integration + 2 * f(k)
        else:
            integration = integration + 3 * f(k)
    
    # Finding final integration value
    integration = integration * 3 * h / 8
    
    return integration
    
# Input section
lower_limit = 0
upper_limit = np.pi*2
sub_interval = 20

# Call trapezoidal() method and get result
result = simpson38(lower_limit, upper_limit, sub_interval)
print("Integration result by Simpson's 3/8 method is: %0.6f" % (result) )
```
